```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(1 , 11) ; 
y = 2 * x ; 

plt.plot(x , y, color = 'r' , linestyle = '--' , linewidth = 2) ; 
plt.title("Line plot")
plt.xlabel("x-lavel")
plt.ylabel("y-lavel")
plt.show() 
```

```python
import numpy as np  
from matplotlib import pyplot as plt 

x = np.arange(1 , 20 , 2) ; 
y = x * x ; 
y2 = 8*x ; 
plt.plot(x , y , color = 'b' , linewidth = 4)
plt.plot(x , y2 , color = 'r', linewidth = 4)
plt.grid(True)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Parabola")
plt.show()
```

```python
import numpy as np  
from matplotlib import pyplot as plt 

x = np.arange(1,11)
y1 = 2 * x 
y2 = 3 * x


plt.subplot(2,1,1)
plt.plot(x , y1 , color = 'r' , linewidth = 2)
plt.grid

plt.subplot(2,1,2)
plt.plot(x,y2,color='b',linewidth = 2)

plt.grid(True)
plt.show()
```

```python
import numpy as np 
from matplotlib import pyplot as plt

student  = { 'Sachin' : 10 , 'Roshan' : 20 , 'Rahul' : 40}

names = list(student.keys())
values = list(student.values())

plt.bar(names, values)
plt.title("Bar chart")
plt.grid(True)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import numpy as np 
from matplotlib import pyplot as plt

student  = { 'Sachin' : 10 , 'Roshan' : 20 , 'Rahul' : 40}

names = list(student.keys())
values = list(student.values())

plt.barh(names, values)
plt.title("Bar chart")
plt.grid(True)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import numpy as np 
from matplotlib import pyplot as plt

x = [10, 20, 30 , 40 , 50]
y = [10 , 2 , 30 , 4 , 50 ]
z = [20,3,1,30,50]

plt.scatter(x,y , marker="*", c = 'g' , s = 100)
plt.scatter(x, z , marker = "o" , c = 'b', s = 100 )

plt.show()
```

```python
import numpy as np 
from matplotlib import pyplot as plt

x = [10, 20, 30 , 40 , 50]
y = [10 , 2 , 30 , 4 , 50 ]
z = [20,3,1,30,50]

plt.subplot(1,2,1)
plt.scatter(x,y , marker="*", c = 'g' , s = 100)

plt.subplot(1,2,2)
plt.scatter(x, z , marker = "o" , c = 'b', s = 100 )

plt.show()
```
